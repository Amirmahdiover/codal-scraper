```python
import requests
import pandas as pd

# API URL
url = "https://founda.boors365.ir/api/v1/tickers?force_refresh=false&max_concurrent=20"

# Send GET request to the API
response = requests.get(url)

# Check for successful response
if response.status_code == 200:
    data = response.json()
    
    # Extract 'name' from each item in 'data'
    company_names = [item['full_name'] for item in data.get('data', [])]
    tickers =[item['name'] for item in data.get('data',[])]
    # Create a DataFrame
    df = pd.DataFrame({
        'ticker': tickers,
        'company_name': company_names
    })
    
```
